```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

x = np.random.rand(500)
y = np.dot(x,9)  +  np.random.rand(500)

data = pd.DataFrame({'x': x , 'y': y})
data.head()
```

```python
plt.figure(figsize=(12,10))
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Data")
plt.show()
```

```python
def gradient_descent(x,y):
    n = len(x)
    m = b = 0
    epochs = 1000
    learning_rate = 0.08

    for i in range(epochs):
        y_predicted = np.dot(m, x) + b 
        d_m = -(2/n) * np.dot(x , (y - y_predicted))  
        d_b = -(2/n) * sum(y - y_predicted)
        m = m - learning_rate * d_m
        b = b - learning_rate * d_b

    y_predicted = np.dot(m, x) + b

    plt.figure(figsize = (12,10))
    plt.scatter(x, y)
    plt.plot(x, y_predicted, 'r', linewidth = 3)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Linear regression")
    plt.show()
    
gradient_descent(x,y)
```
