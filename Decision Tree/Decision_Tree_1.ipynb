```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
dataset = pd.DataFrame(data = data['data'],columns=data['feature_names'])
dataset
```

```python
data.target_names
```

```python
from sklearn.model_selection import train_test_split

X = dataset.copy()
y = data['target']

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
```

```python
clf.get_params()
```

```python
prediction = clf.predict(X_test)
prediction
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, prediction, labels=[0,1])
```

```python
from sklearn.metrics import precision_score,recall_score
precision_score = precision_score(y_test,prediction)
recall_score = recall_score(y_test,prediction)
print("The precision is : ", precision_score)
print("The recall is : ", recall_score)
```

```python
from sklearn.metrics import classification_report
classification_report = classification_report(y_test,prediction,target_names = ['malignant', 'benign'])
print(classification_report)
```

```python
feature_importances = pd.DataFrame(data = clf.feature_importances_,index = data.feature_names, columns=["Importance"]).sort_values(by="Importance", ascending=False)
feature_importances
```

```python
feature_importances.head(12).plot(kind = 'bar')
plt.show()
```
